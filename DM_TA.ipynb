```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Pasar Grogol.csv')
print('Ukuran dataset: %d baris dan %d kolom' % dataset.shape)
print('Lima data teratas:')
print(dataset.head())
```

```python
dataset.dtypes
```

```python
# Penambahan Kolom Order Month
import datetime
dataset['order_month'] = dataset['tanggal'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
print(dataset.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(dataset['order_month'], dataset['harga'])
plt.show()
```
