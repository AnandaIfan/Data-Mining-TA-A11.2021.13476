```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Pasar Grogol.csv')
print('Ukuran dataset: %d baris dan %d kolom' % dataset.shape)
print('Lima data teratas:')
print(dataset.head())
```

```python
dataset.dtypes
```

```python
# Penambahan Kolom Order Month
import datetime
dataset['order_month'] = dataset['tanggal'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
print(dataset.head())
```

```python
# Membuat Line Chart Trend Pertumbuhan GMV
import matplotlib.pyplot as plt
plt.plot(dataset['order_month'], dataset['harga'])
plt.show()
```

```python
# Title and Axis Labels
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['harga'].sum().plot()
plt.xlabel('Order Month')
plt.ylabel('Total GMV')
plt.show()
```

```python
# Kustomisasi Line dan Point
plt.figure(figsize=(15, 5))
dataset.groupby(['order_month'])['harga'].sum().plot(color='green', marker='o', linestyle='-.', linewidth=2)
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount', fontsize=15)
plt.show()
```
